```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
# Assuming your data is i888
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
df.shape
```

```python
df.shape
```

```python
df.head(20)
```

```python
df.info()
```

```python
# Drop rows with missing values in 'Amount' column
df = df.dropna(subset=['Amount'])

# Fill missing values with the mean of the 'Amount' column
df['Amount'].fillna(df['Amount'].mean(), inplace=True)
```

```python
df.info()
```

```python
# Drop unnecessary columns
df = df.drop(['Status', 'unnamed1'], axis=1)
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
# Remove duplicate rows
df = df.drop_duplicates()
```

```python
df.columns
```

```python
df.describe()
```

```python
# Gender
```

```python
# Assuming df is your DataFrame
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))

# Countplot for 'Gender'
ax = sns.countplot(x='Gender', data=df)

# Add count values on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Assuming df is your DataFrame
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

# Countplot for 'Gender' with hue='Age Group'
sns.countplot(x='Gender', hue='Age Group', data=df)

plt.title('Distribution of Gender by Age Group')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Age Group', loc='upper right')
plt.show()
```

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Assuming df is your DataFrame
state_amount_sum = df.groupby('State')['Amount'].sum().reset_index()

# Display the result
print(state_amount_sum)
```

```python
# Assuming df is your DataFrame
state_amount_sum = df.groupby('State')['Amount'].sum().reset_index()

# Sort the DataFrame by 'Amount' in descending order
state_amount_sum = state_amount_sum.sort_values(by='Amount', ascending=False)

# Set the default figure size for Seaborn plots and rotate x-axis labels
sns.set(rc={'figure.figsize': (15, 5)})
plt.xticks(rotation=45)

# Bar plot for 'State' and 'Amount' with custom order
sns.barplot(x='State', y='Amount', data=state_amount_sum, order=state_amount_sum['State'])
plt.title('Total Amount by State')
plt.xlabel('State')
plt.ylabel('Total Amount')
plt.show()
```

```python
# Assuming df is your DataFrame
state_total_orders = df.groupby('State')['Orders'].sum().reset_index()

# Set the default figure size for Seaborn plots and rotate x-axis labels
sns.set(rc={'figure.figsize': (15, 5)})
plt.xticks(rotation=45)

# Bar plot for 'State' and 'Total Orders'
sns.barplot(x='State', y='Orders', data=state_total_orders)
plt.title('Total Number of Orders by State')
plt.xlabel('State')
plt.ylabel('Total Orders')
plt.show()
```

```python
# Assuming df is your DataFrame
marital_status_total_orders = df.groupby('Marital_Status')['Orders'].sum().reset_index()

sns.set(rc={'figure.figsize': (8, 6)})
sns.barplot(x='Marital_Status', y='Orders', data=marital_status_total_orders)
plt.title('Total Number of Orders by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Total Orders')
plt.show()
```

```python
# Assuming df is your DataFrame
sns.set(rc={'figure.figsize': (15, 5)})
plt.xticks(rotation=45)
sns.countplot(x='Occupation', data=df)
plt.title('Count of Individuals by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.show()
```

```python
# Assuming df is your DataFrame
product_category_avg_amount = df.groupby('Product_Category')['Amount'].mean().reset_index()

sns.set(rc={'figure.figsize': (15, 5)})
plt.xticks(rotation=45)
sns.barplot(x='Product_Category', y='Amount', data=product_category_avg_amount)
plt.title('Average Amount by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Average Amount')
plt.show()
```
